```python
from bs4 import BeautifulSoup

import requests
```

```python
page = requests.get('https://www.nobroker.in/property/sale/mumbai/multiple?searchParam=W3sibGF0IjoxOC45OTg2NDA2LCJsb24iOjcyLjgxNzM1OTksInBsYWNlSWQiOiJDaElKeFl3ME5KZk81enNSN0FCazRLcXJkeDgiLCJwbGFjZU5hbWUiOiJXb3JsaSJ9LHsibGF0IjoxOS4wNTk1NTk2LCJsb24iOjcyLjgyOTUyODcsInBsYWNlSWQiOiJDaElKZTlMNEktSEk1enNSZUdvam1yU1dlVU0iLCJwbGFjZU5hbWUiOiJCYW5kcmEgV2VzdCJ9LHsibGF0IjoxOS4xMDc1NDI0LCJsb24iOjcyLjgyNjMxNDIsInBsYWNlSWQiOiJDaElKQjdyVWxiX0o1enNSb2M2Mk5weDhCRFUiLCJwbGFjZU5hbWUiOiJKdWh1In1d&radius=2.0&city=mumbai&locality=Worli,&locality=Bandra%20West,&locality=Juhu')

page
```

```python
soup = BeautifulSoup(page.content)

soup
```

```python
list_of_article = soup.find_all('article')

title = []
price = []
emi = []
location = []
area = []
for article in list_of_article:
    title.append(article.find('h2').text)
    area.append(article.find('h2').text.split('In')[-1])
    
    temp = article.find_all('div', class_ = "font-semi-bold heading-6")
    price.append(temp[2].text)
    emi.append(temp[1].text)

    temp = article.find_all('div', class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95")
    location.append([j.text for j in temp][0])
```

```python
import pandas as pd

df = pd.DataFrame(data = {'Title' : title, "Location" : location, 'Area' : area, 'Price' : price, 'EMI' : emi})
```
